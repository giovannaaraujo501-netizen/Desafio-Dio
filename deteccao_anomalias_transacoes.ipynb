```python

# Instalar bibliotecas
!pip install pandas scikit-learn

# Importar bibliotecas
import pandas as pd
from sklearn.ensemble import IsolationForest

# Criando dados de transações
dados = pd.DataFrame({
    'valor': [100, 120, 110, 115, 105, 5000, 108, 112, 98, 7000]
})

# Criando modelo
modelo = IsolationForest(contamination=0.2, random_state=42)

# Treinando modelo
modelo.fit(dados)

# Detectando anomalias
dados['resultado'] = modelo.predict(dados)

# Convertendo resultado
dados['resultado'] = dados['resultado'].map({
    1: 'Normal',
    -1: 'Anomalia'
})

# Mostrar resultados
print(dados)
```

Sistema de Detecção de Anomalias em Transações FinanceirasObjetivo
Este projeto foi desenvolvido em Python com o objetivo de identificar transações financeiras suspeitas utilizando técnicas de Machine Learning.
Descrição
O sistema analisa valores de transações e detecta comportamentos fora do padrão, classificando possíveis anomalias automaticamente.
Tecnologias UtilizadasPythonPandasScikit-learnMatplotlibAlgoritmo Utilizado
Foi utilizado o algoritmo Isolation Forest, muito aplicado em problemas de detecção de fraudes e anomalias.
Funcionamento
O modelo analisa os dados das transações e identifica valores incomuns em comparação ao comportamento normal.
Resultado
O sistema conseguiu identificar transações suspeitas automaticamente, demonstrando como Machine Learning pode auxiliar na prevenção de fraudes financeiras.

```python

import matplotlib.pyplot as plt

# Separando dados
normais = dados[dados['resultado'] == 'Normal']
anomalias = dados[dados['resultado'] == 'Anomalia']

# Criando gráfico
plt.figure(figsize=(10,5))

plt.scatter(normais.index, normais['valor'], label='Normal')
plt.scatter(anomalias.index, anomalias['valor'], label='Anomalia')

plt.xlabel('Transação')
plt.ylabel('Valor')
plt.title('Detecção de Anomalias')

plt.legend()

plt.show()
```

Análise dos Resultados
O gráfico mostra o comportamento das transações financeiras analisadas.
As transações classificadas como "Anomalia" representam valores fora do padrão esperado pelo modelo de Machine Learning.
Isso demonstra como algoritmos podem auxiliar na identificação de possíveis fraudes ou movimentações suspeitas.
